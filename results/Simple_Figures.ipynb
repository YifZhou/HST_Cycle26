```python
%matplotlib notebook
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette(sns.color_palette("deep"))
```

```python
df = pd.read_csv('./Cycle26_Summary_public.csv')
```

# some plots

```python
fig = plt.figure()
plt_country = df['PI Country'].value_counts().plot(kind='pie')
```

```python
fig = plt.figure(figsize=(8, 4))
category_Nprop = df['Science Category'].value_counts().sort_values()
category_Nprop.plot(kind='barh', title='No. of Proposal by Category')
ax2 = fig.add_subplot(212)
category_Nprop.plot(kind='pie', labels=[""]*10)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 4))
scigp = df.groupby('Science Category')['Orbits'].sum().sort_values()
scigp.plot(kind='barh', title='N of awarded orbits')
fig.add_subplot(212)
scigp.plot(kind='pie', labels=['']*10)
fig.tight_layout()
fig = plt.figure(figsize=(8, 4))
scigp_mean = df.groupby('Science Category')['Orbits'].mean().sort_values()
scigp_mean.plot(kind='barh', title='Average N Orbits per Program')
fig.add_subplot(212)
scigp_mean.plot(kind='pie', labels=['']*10)
fig.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_subplot(211)
sns.distplot(df['Orbits'])
ax.set_title('Number of Orbit Distribution')
ax = fig.add_subplot(212)
sns.distplot(df.loc[(df['Orbits']>=35) & (df['Orbits'] < 50), 'Orbits'], bins=15)
ax.set_title('Number of Orbit Distribution 20<N<50')
fig.tight_layout()
```

```python
plt.figure()
sns.distplot(df['N People'])
plt.title('Number of Investigators Distributions')
```

```python
sns.jointplot(x='N People', y='Orbits', data=df, kind='kde', 
              xlim=[0, 25], ylim=[0, 200])
```
